## Setup

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

## Webscraping with requests

```python
# Website: FiveThirtyEight
url = "https://edition.cnn.com/2022/11/17/media/twitter-exodus-reliable-sources/index.html"
```

```python
html = requests.get(url)
```

```python
print(f"HTTP status code: {html.status_code}")
```

```python
soup = BeautifulSoup(html.text, 'html.parser')
```

```python
print(soup.get_text())
```

```python
# Titel des Artikels
article_title = soup.h1.text.replace('\n', '')
print(article_title)
```

```python
# Thema des Artikels
article_topic = soup.find("div", {"class":"source inline-placeholder"}).text.replace('\n', '')
print(article_topic)
```

```python
# Autor des Artikels
article_author = soup.find("a",{"class":"byline__link"}).text
print(article_author)
```

```python
# Veröffentlichungszeitpunkt des Artikels
article_datetime = soup.find("div",{"class":"timestamp"}).text.replace("\n","").replace("\t","")
print(article_datetime)
```

```python
# Artikeltext
article_text_html = soup.findAll("div", {"class":"article__content"})
```

```python
article_text_list = []
for i in article_text_html:
    article_text_list.append(i.get_text().replace("\xa0",""))
```

```python
article_text = "".join(article_text_list)
article_text = article_text.replace("\n","")
```

```python
article_text
```

```python
# Erstellung des Dataframes
df = pd.DataFrame({
    "Titel" : article_title,
    "URL" : url,
    "Thema" : article_topic,
    "Autor" : article_author,
    "Datum" : article_datetime,
    "Artikeltext" : article_text,
}, index=[0])
```
